## Basics of Matplotlib

### Make a simple plot

```python
# list of height
height = [140, 150, 120, 193]
# List of weight
weight = [70, 80, 90, 100]

# Draw the plot
plt.plot(height, weight)
```

- We pass two lists as our input arguments to **plot()** method and invoke the required plot.
- Here note that the first array appears on the x-axis and second array appears on the y-axis of the plot.

### Title, Labels, and Legends
Add the title, and name x-axis and y-axis using methods **title()**, **xlabel()** and **ylabel()** respectively.

```python
# Draw the plot
plt.plot(height,weight)

# add title
plt.title('Relationship between height and weight')
# label x axis
plt.xlabel('Height')
# label y axis
plt.ylabel('Weight')
```

```python
# List of calories_burnt
calories_burnt = [65, 75, 95, 99]

# Draw the plot for calories burnt
plt.plot(calories_burnt)
# draw the plot for weight
plt.plot(weight)
# draw the plot for height
plt.plot(height)
```

You can use the `legend()` which takes **labels** and **loc** as label names and location of legend in the figure as paremeters.

```python
# Draw the plot for calories burnt
plt.plot(calories_burnt)
# draw the plot for weight
plt.plot(weight)
# draw the plot for height
plt.plot(height)

# add legend in the lower right part of the figure
plt.legend(labels = ['Calories Burnt', 'Weight', 'Height'], loc='lower right')
```

- Look at the X axis, can we add labels to show that each belong to different persons?
- The labeled values on any axis is known as a **tick**. You can use the `xticks` to change both the location of each tick and it's label.

```python
# draw the plot
plt.plot(calories_burnt)
plt.plot(weight)
plt.plot(height)

# Add legend in the lower right part of the figure
plt.legend(labels=['Calories Burnt', 'Weight', 'Height'], loc='lower right')

# Set labels for each of these persons
plt.xticks(ticks=[0,1,2,3], labels = ['p1', 'p2', 'p3', 'p4'])
```

### Size, Colors, Markers and Line styles

- You can also specify the size of the figure using method **`figure()`** and passing the values as a tuple of the length of rows and columns to the argument figsize. 
- The values of length are considered to be in **inches**.

```python
# figure size in inches
plt.figure(figsize = (12,5))

# Draw the plot
plt.plot(calories_burnt, marker= 'o')
plt.plot(weight, marker = '*')
plt.plot(height, marker = 'x')

# Add legend in the lower right part of the figure
plt.legend(labels=['Calories Burnt', 'Weight', 'Height'], loc='lower right')

# set labels for each of these persons
plt.xticks(ticks=[0,1,2,3], labels=['p1', 'p2', 'p3', 'p4']);
```

- With every X and Y argument, you can also pass an optional third argument in the form of a string which indicates the colour and line type of the plot. 
- The default format is `b-` which means a **solid blue line**. In the figure below we use `go` which means **green circles**.

```python
# Draw the plot
plt.plot(calories_burnt, marker= 'o')
plt.plot(weight,'y--', marker='*')
plt.plot(height, marker = 'x')

# add legend in the lower right part of the figure
plt.legend(labels = ['Calories Burnt', 'Weight', 'Height'], loc = 'lower right')

# set labels for each of these persons
plt.xticks(ticks=[0,1,2,3], labels=['p1', 'p2', 'p3', 'p4']);
```

### Figure and subplots

- We can use **`subplots()`** method to add more than one plots in one figure. It takes two arguments: **nrows, ncols** which indicate rows and columns respectively. 
- This method creates two objects: **figure** and **axes** which we store in variables `fig` and `ax`. 

```python
# Create 3 plots
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15,4))

# Plot on 0 row and 0 column
ax[0].plot(calories_burnt, 'go')

# Plot on 0 row and 1 column
ax[1].plot(weight, 'y--', marker = '*')

# Plot on 0 row and 2 column
ax[2].plot(height, 'y--', marker = 'o')

# Set titles for subplots
ax[0].set_title('Calories Burnt')
ax[1].set_title('Weight')
ax[2].set_title('Height')

# set ticks for each of these persons
ax[0].set_xticks(ticks=[0,1,2,3])
ax[1].set_xticks(ticks=[0,1,2,3])
ax[2].set_xticks(ticks=[0,1,2,3])

# set labels for each of these persons
ax[0].set_xticklabels(labels = ['p1', 'p2', 'p3', 'p4'])
ax[1].set_xticklabels(labels = ['p1', 'p2', 'p3', 'p4'])
ax[2].set_xticklabels(labels = ['p1', 'p2', 'p3', 'p4'])
```

There will be times when you need to have a common axis for all your subplots. You can do this by using the **sharex** and **sharey** paremeters of `subplot()`.

```python
# Create 3 plots
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(14,4), sharex = True, sharey = True)

# Plot on 0 row and 0 column
ax[0].plot(calories_burnt, 'y--', marker = 'x')

# Plot on 0 row and 1 column
ax[1].plot(weight, 'y--', marker = '*')

# Plot on 0 row and 2 column
ax[2].plot(height, 'y--', marker = 'o')

# Set titles for subplots
ax[0].set_title('Calories Burnt')
ax[1].set_title('Weight')
ax[2].set_title('Height')

# set ticks for each of these persons
ax[0].set_xticks(ticks=[0,1,2,3])
ax[1].set_xticks(ticks=[0,1,2,3])
ax[2].set_xticks(ticks=[0,1,2,3])

# set labels for each of these persons
ax[0].set_xticklabels(labels = ['p1', 'p2', 'p3', 'p4'])
ax[1].set_xticklabels(labels = ['p1', 'p2', 'p3', 'p4'])
ax[2].set_xticklabels(labels = ['p1', 'p2', 'p3', 'p4'])
```

Now both x and y axes are only labelled once for each of the outer plots. This is because the inner plots "share" both the axes.

## Working with real data

```python
# Read the dataset
data = pd.read_csv('dataset/bigmart_data.csv')

# drop the null values
data = data.dropna(how = 'any')

# view the top results
data.head()
```

### Line Chart

- We will create a line chart to denote the **mean price per item**.
- In matplotlib, **line chart** is the default plot when using the `plot()`. 

```python
mean_price = data.groupby('Outlet_Establishment_Year')['Item_Outlet_Sales'].mean()
print(mean_price)

# Printing values and index
print(mean_price.index, mean_price.values)
```

```python
# Mean price based on item type
mean_price = data.groupby('Item_Type')['Item_MRP'].mean()[:10]

x = mean_price.index.tolist()
y = mean_price.values.tolist()

# set figure size
plt.figure(figsize=(12, 4))

# Set title
plt.title('Mean price for each item type')

# set axis labels
plt.xlabel('Item Type')
plt.ylabel('Mean Price')

# set xticks and rotation = 45
plt.xticks(labels = x, ticks = np.arange(len(x)), rotation = 45)

plt.plot(x, y, marker = 'o');
```

### Bar Chart

- We want to have a look at **the mean sales for each outlet type**. A bar chart is another simple type of visualization that is used for categorical variables.
- You can use `plt.bar()` instead of `plt.plot()` to create a bar chart.

```python
# sales by outlet size
mean_sales = data.groupby('Outlet_Size')['Item_Outlet_Sales'].mean()

# sort by sales
mean_sales.sort_values(inplace=True)

x = mean_sales.index.tolist()
y = mean_sales.values.tolist()

plt.figure(figsize = (4,4))

# set axis labels
plt.xlabel('Outlet Size')
plt.ylabel('Sales')

# set title
plt.title('Mean sales for each outlet type')

# set xticks 
plt.xticks(labels = x, ticks = np.arange(len(x)))

plt.bar(x, y, color = ['red', 'orange', 'magenta']);
```

### Histogram

- We want to check the **Distribution of Item price**. Histograms are a very common type of plots when we are looking at data like height and weight, stock prices, waiting time for a customer, etc which are continuous in nature. Histograms data is plotted within a range against its frequency. 
- Histograms are very commonly occurring graphs in probability and statistics and form the basis for various distributions like the normal -distribution, t-distribution, etc.
- You can use `plt.hist()` to draw a histogram.

```python
# title
plt.title('Item price distribution')

# labels
plt.xlabel('Item_MRP')
plt.ylabel('Frequency')

# plot histogram
plt.hist(data['Item_MRP'], bins=20, color = 'lightblue')
```

### Box Plots

- We want to check the **Distribution of sales**. Box plot shows the three quartile values of the distribution along with extreme values. 
- The whiskers extend to points that lie within 1.5 IQRs of the lower and upper quartile, and then observations that fall outside this range are displayed independently. This means that each value in the boxplot corresponds to an actual observation in the data.

```python
# Create sales data
sales_data = data[['Item_Outlet_Sales']]

# create outlier point shape
red_diamond = dict(markerfacecolor='r', marker='D')

# Set title
plt.title('Item Sales distribution')

# Make the boxplot
plt.boxplot(sales_data.values, labels=['Item Sales'], flierprops=red_diamond);
```

- You can also create multiple boxplots for different columns of your dataset.
- In order to plot multiple boxplots, you can use the same `subplots()`.

```python
combined_data = data[['Item_Weight', 'Item_MRP']]

# create outlier point shape
red_diamond = dict(markerfacecolor='r', marker='D')

# generate subplots
fig, ax = plt.subplots()

# make the boxplot
plt.boxplot(combined_data.values, labels = ['Item_Weight', 'Item_MRP'], flierprops=red_diamond);
```

### Violin Plots

Find the **Density distribution of Item weights and Item price**.

### Scatter Plots
