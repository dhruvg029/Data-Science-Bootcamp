## Import libraries and load images

```python
# Importing modules from skimage to read and display images
import skimage
from skimage.io import imshow, imread, imsave
from skimage.transform import resize
import numpy as np
import pandas as pd
import cv2

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

print('skimage version:', skimage.__version__)
```

```python
# Reading the image using its path
image = imread('images/dog.jpg')

# Plotting the image
imshow(image)
```

## Operations on Images

### Resizing of an image

Image resizing refers to the scaling of images. Scaling comes in handy in many image processing as well as machine learning applications. 

**Syntax:** **`cv2.resize(source, dsize, dest, fx, fy, interpolation)`**

```python
# Resizing of an image
half_size = cv2.resize(image, (0, 0), fx = 0.1, fy = 0.1)

# Plot the image
imshow(half_size)
```

```python
# Calculate the number of pixels
print('Number of Pixels in original image:', image.shape[0] * image.shape[1])
print('Number of Pixels in resized image:', half_size.shape[0] * half_size.shape[1])
```

### Rotation of an image

You can use **`imutils.rotate()`** method is used to rotate an image in any degree.

```python
# Rotation of the image
import imutils

rot1 = imutils.rotate(image, angle = 90)
rot2 = imutils.rotate(image, angle = 180)
rot3 = imutils.rotate(image, angle = 270)

titles = ['Original', '90 degrees', '180 degrees', '270 degrees']
images = [image, rot1, rot2, rot3]
count = 4
 
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(titles[i])
    plt.imshow(images[i])
 
plt.show()
```

### Translating an image

Translation refers to the rectilinear shift of an object i.e. an image from one location to another.

```python
# Translating an image
height, width = image.shape[:2]
half_height, half_width = height / 2, width / 2

T = np.float32([[1, 0, half_width], [0, 1, half_height]])
img_translation = cv2.warpAffine(image, T, (width, height))

Titles = ['Original', 'Half']
images = [image, img_translation]
count = 2
 
for i in range(count):
    plt.subplot(1, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])
 
plt.show()
```

### Negative of an image

When we try to negatively transform an image, the brightest areas are transformed into the darkest and the darkest areas are transformed into the brightest.

```python
# Negative of an image
L = 2^8          
neg = (L - 1) - image

titles = ['Original image', 'Negative image']
images = [image, neg]
count = 2
 
for i in range(count):
    plt.subplot(1, 2, i + 1)
    plt.title(titles[i])
    plt.imshow(images[i])
 
plt.show()
```

```python
# Negative of a grayscale image
gray_image = imread('images/grayscale_image.jpg')

L = 2^8          
neg = (L - 1) - gray_image

titles = ['Original image', 'Negative image']
images = [gray_image, neg]
count = 2
 
for i in range(count):
    plt.subplot(1, 2, i + 1)
    plt.title(titles[i])
    plt.imshow(images[i])
 
plt.show()
```

### Cropping of an image

Cropping an image is one of the most basic image operations. Use the slicing method to perform cropping.

```python
# Image cropping
cropped_image1 = image[50:200, 150:200]
cropped_image2 = image[150:200, 30:90]
cropped_image3 = image[80:180, 90:100]

titles = ['Original', 'Cropped', 'Cropped', 'Cropped']
images = [image, cropped_image1, cropped_image2, cropped_image3]
count = 4
 
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(titles[i])
    plt.imshow(images[i])
 
plt.show()
```

### Contours of an image

Contours can be explained simply as a curve joining all the continuous points (along the boundary), having same color or intensity. The contours are a useful tool for shape analysis and object detection and recognition.

```python
# Contours of a grayscale image
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

edged = cv2.Canny(grayscale_image, 30, 200)
contours, hierarchy = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

imshow(edged)
```

```python
# Contours of a colored image
cv2.drawContours(image, contours, -1, (0, 255, 0), 3)

# Check the image
imshow(image)
```

```python
# Colored image to HSV
hsvImage = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Check the image
imshow(hsvImage)
```

### Blurring of an image

Blur the image with the help of different blurring techniques like **`cv2.medianBlur()`**, **`cv2.GaussianBlur()`** and **`cv2.blur()`**.

```python
img = imread('images/dog.jpg')

# USE THE MEDIAN BLURRING TECHNIQUE
median = cv2.medianBlur(img, 5)
# compare = np.concatenate((img, median), axis=1)

median1 = cv2.medianBlur(img, 3)
# compare1 = np.concatenate((img, median1), axis=1)

median2 = cv2.medianBlur(img, 7)
# compare2 = np.concatenate((img, median2), axis=1)

titles = ['Original image', 'Image 1', 'Image 2', 'Image 3']
images = [img, median, median1, median2]
count = 4
 
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(titles[i])
    plt.imshow(images[i])
 
plt.show()
```
