# Conversion of an image into different formats 

## Import libraries

```python
# Importing modules from skimage to read and display images
import skimage
from skimage.io import imshow, imread, imsave
from skimage.color import rgb2gray, rgba2rgb
import numpy as np
import pandas as pd
import cv2

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

print('skimage version:', skimage.__version__)
```

## Loading an RGB image

```python
# reading the image using its path
image = imread('images/sample_image.jpg')

# Checking shape of image
print('Shape of the RGB image:', image.shape)

# Plotting the image
imshow(image)
```

## Loading RGBA image

```python
# loading image using imread
image_rgba = imread('images/mid_alpha.png')

# Checking shape of image
print('Shape of the RGBA image:', image_rgba.shape)

# Plotting the image
imshow(image_rgba)
```

## Converting RGBA to RGB

```python
# converting rgba to rgb
image_rgb = rgba2rgb(image_rgba)

# Checking shape of image
print('Shape of the RGB image:', image_rgb.shape)

# Plot the image
imshow(image_rgb)
```

## Converting RGB to grayscale image

```python
# converting rgb to gray
image_gray = rgb2gray(image)

# Checking shape of image
print('Shape of the grayscale image:', image_gray.shape)

# Plotting the image
imshow(image_gray)
```

## Loading RGB in grayscale format 

```python
# Load the image
image_ = imread('images/sample_image.jpg', as_gray=True)

# Show the image
imshow(image_)
```
