# Different thresholding techniques

## Importing libraries and loading image

```python
# Importing the required libraries
from skimage.io import imread, imshow
import cv2

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading the image and plotting it
image = imread('images/dog.jpg')
imshow(image, cmap = 'gray')
```

## Different thresholding techniques

**Global thresholding:** A global thresholding technique is one which makes use of a single threshold value for the whole image, whereas local thresholding technique makes use of unique threshold values for the partitioned subimages obtained from the whole image.
**Adaptive thresholding:**
Adaptive thresholding is a form of image thresholding technique in which rather than specifying the threshold value manually or using any restrictions, the threshold value is adjusted and selected automatically according to the image pixels and layout for converting the image pixels to grayscale or a binary image.

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Global Thresholding
thresh, dest = cv2.threshold(gray, 75, 255, cv2.THRESH_BINARY)

# Adaptive Mean Thresholding
thresh1 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 199, 5)

# Adaptive Gaussian Thresholding
thresh2 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 199, 5)

# Plotting the edges
titles = ['IMAGE', 'GLOBAL_THRESH', 'ADAPTIVE_THRESH_M', 'ADAPTIVE_THRESH_G']
images = [gray, dest, thresh1, thresh2]
count = 4

plt.figure(figsize = (12, 12))

for i in range(count):
    plt.subplot(1, 4, i + 1)
    plt.title(titles[i])
    plt.imshow(images[i], cmap = 'gray')
 
plt.show()
```
