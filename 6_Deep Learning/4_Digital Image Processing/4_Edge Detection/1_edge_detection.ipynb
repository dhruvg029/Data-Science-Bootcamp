# Edge detection in images

Edge detection is one of the fundamental image-processing tasks used in various Computer Vision tasks to identify the boundary or sharp changes in the pixel intensity. It plays a crucial role in object detection, image segmentation and feature extraction from the image. In Real-time edge detection, the image frame coming from a live webcam or video is continuously captured and each frame is processed by edge detection algorithms which identify and highlight these edges continuously.

## Loading libraries and importing image

```python
# Importing the required libraries
from skimage.io import imread, imshow
from skimage.filters import sobel_h, prewitt_h, prewitt_v, sobel_v, roberts, laplace
```

```python
# Reading the image and plotting it
image = imread('images/dog.jpg', as_gray = True)
imshow(image, cmap='gray')
```

```python
# Checking the shape of the image
print('Shape:', image.shape)
```

## Different types of kernels

### Sobel kernel

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Calculating the horizontal edges using sobel kernel
edges_sobel_horizontal = sobel_h(image)

# Calculating the vertical edges using soble kernel
edges_sobel_vertical = sobel_v(image)

# Plotting the edges
titles = ['IMAGE', 'HORIZONTAL SOBEL', 'VERTICAL SOBEL']
images = [image, edges_sobel_horizontal, edges_sobel_vertical]
count = 3

plt.figure(figsize = (10, 10))

for i in range(count):
    plt.subplot(1, 3, i + 1)
    plt.title(titles[i])
    plt.imshow(images[i], cmap = 'gray')
 
plt.show()
```

### Prewitt Kernel

```python
# Calculating horizontal edges using prewitt kernel
edges_prewitt_horizontal = prewitt_h(image)

# Calculating vertical edges using prewitt kernel
edges_prewitt_vertical = prewitt_v(image)

# Plotting the edges
titles = ['IMAGE', 'HORIZONTAL PREWITT', 'VERTICAL PREWITT']
images = [image, edges_prewitt_horizontal, edges_prewitt_vertical]
count = 3

plt.figure(figsize = (10, 10))

for i in range(count):
    plt.subplot(1, 3, i + 1)
    plt.title(titles[i])
    plt.imshow(images[i], cmap = 'gray')
 
plt.show()
```

### Roberts operator

```python
# Calculating edges using roberts operator
edges_roberts = roberts(image)

# Plotting the edges
titles = ['IMAGE', 'ROBERTS']
images = [image, edges_roberts]
count = 2

plt.figure(figsize = (10, 10))

for i in range(count):
    plt.subplot(1, 2, i + 1)
    plt.title(titles[i])
    plt.imshow(images[i], cmap = 'gray')
 
plt.show()
```

### Laplacian operator

```python
# Calculating edges using laplacian operator
edges_laplace = laplace(image)

# Plotting the edges
titles = ['IMAGE', 'LAPLACE']
images = [image, edges_laplace]
count = 2

plt.figure(figsize = (10, 10))

for i in range(count):
    plt.subplot(1, 2, i + 1)
    plt.title(titles[i])
    plt.imshow(images[i], cmap = 'gray')
 
plt.show()
```

### Harris Corner detection

```python
import cv2
import numpy as np

# Reading the image and plotting it
image = imread('images/dog.jpg')

operatedImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
operatedImage = np.float32(operatedImage)

# Apply Harris Corner detection algorithm
dest = cv2.cornerHarris(operatedImage, 2, 5, 0.07)
 
# Results are marked through the dilated corners
dest = cv2.dilate(dest, None)
 
# Reverting back to the original image,
# with optimal threshold value
image[dest > 0.01 * dest.max()] = [0, 0, 255]
 
# the window showing output image with corners
imshow(image)
```
