# Image thresholding

Thresholding is one of the segmentation techniques that generates a binary image (a binary image is one whose pixels have only two values â€“ 0 and 1 and thus requires only one bit to store pixel intensity) from a given grayscale image by separating it into two regions based on a threshold value. Hence pixels having intensity values greater than the said threshold will be treated as white or 1 in the output image and the others will be black or 0. 

You can read more about the image thresholding techniques from the article: [Image Thresholding](https://www.geeksforgeeks.org/thresholding-based-image-segmentation/).

## Importing libraries and loading image

```python
# Importing modules from skimage to read and display images
import skimage
from skimage.io import imshow, imread, imsave
from skimage.transform import resize
import numpy as np
import pandas as pd
import cv2

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

print('skimage version:', skimage.__version__)
```

```python
# Read the image
image = imread('images/dog.jpg')

# Load the image
imshow(image)
```

```python
# Converting the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# See the image
imshow(gray_image)
```

## Different image thresholding techniques

```python
# Different image threshold techniques processed
ret1, thresh_img1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
ret2, thresh_img2 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
ret3, thresh_img3 = cv2.threshold(image, 127, 255, cv2.THRESH_TRUNC)
ret4, thresh_img4 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO)
ret5, thresh_img5 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO_INV)

titles = ['IMAGE', 'THRESH_BINARY', 'THRESH_BINARY_INV', 'THRESH_TRUNC', 'THRESH_TOZERO', 'THRESH_TOZERO_INV']
images = [image, thresh_img1, thresh_img2, thresh_img3, thresh_img4, thresh_img5]
count = 6

plt.figure(figsize = (8,8))

for i in range(count):
    plt.subplot(2, 3, i + 1)
    plt.title(titles[i])
    plt.imshow(images[i])
 
plt.show()
```
