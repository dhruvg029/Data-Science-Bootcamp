# Frequency Encoding

- Frequency Encoding assigns a value based on a count of a category's occurrence in the data set.
- This technique should be used with caution, especially when the frequency of the categories is imbalanced.

## Importing and loading data

```python
# Importing necessary libraries
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline

print(pd.__version__)
print(np.__version__)
```

```python
# Loading the data
data = pd.read_csv('datasets/train_bm.csv')

# Check the data
data.head()
```

```python
# Checking the frequency of every item type
data['Item_Type'].value_counts()
```

## Frequency Encoding

```python
# Creating a temporary dataframe of the frequencies
temp = data['Item_Type'].value_counts()

# Creating a new column which will contain the count of all the item types
data['Item_Type_count'] = data['Item_Type'].apply(lambda x: temp[x])

# Checking the data with the new column
data[['Item_Identifier', 'Item_Type', 'Item_Type_count']].head()
```

## Mean Encoding

```python
# Checking the data
data[['Item_Type', 'Item_Outlet_Sales']].head()
```

```python
# Creating a groupby table with the means of the sales related to all the item_types
data.groupby('Item_Type')['Item_Outlet_Sales'].mean()
```

```python
# Performing mean encoding
encoded_mean = data.groupby(['Item_Type'])['Item_Outlet_Sales'].mean().to_dict() 
data['Item_Type_mean'] =  data['Item_Type'].map(encoded_mean) 
  
data.head()
```
